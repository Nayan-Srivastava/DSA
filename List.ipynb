```python
"""
This script finds and prints the minimum value in a given list of integers.
"""

# List of integers to search
y = list(map(int, input("Enter numbers separated by spaces: ").split()))

# Initialize minval with the first element of the list
minval = y[0]

# Iterate through the list to find the minimum value
for i in y:
    # Update minval if a smaller value is found
    if i < minval:
        minval = i

# Print the minimum value found in the list
print("Minimum Value:",minval)
    
```
