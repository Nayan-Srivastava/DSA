#### Decorators
Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

```python
### function copy
### closures
### decorators
```

```python
## function copy means creating a new function that behaves the same as an existing function.
def welcome():
    return "Welcome to the advanced python course"

welcome()
```

```python
wel=welcome # wel is function copy of welcome()
print(wel())
del welcome # delete the original function name
print(wel())
```

```python
##closures functions means nested functions it is a function inside a function,
# used to hide the inner function from outside access.

def main_welcome(msg):
   
    def sub_welcome_method(): # nested function
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome("Welcome everyone")
```

```python
# Passing function as an argument to another function
def main_welcome(func):
   
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func("Welcome everyone to this tutorial")
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome(print) # passing print function as argument
```

```python
def main_welcome(func,lst):
   
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(func(lst))
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome(len,[1,2,3,4,5]) # passing len function as argument with a list
```

```python
len([1,2,3,4,5,6])
```

```python
### Decorator is a special type of closure function that modifies the behavior of the original function.
def main_welcome(func):
   
    def sub_welcome_method(): # nested function
        print("Welcome to the advance python course")
        func() # calling the original function
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
def coure_introduction():
    print("This is an advanced python course")

coure_introduction()
```

```python
main_welcome(coure_introduction)
```

```python
@main_welcome # decorator syntax
def coure_introduction():
    print("This is an advanced python course")
```

```python
## Decorator

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")

```

```python
say_hello()
```

```python
## Decorators With arguments
def repeat(n):
    def decorator(func): # the actual decorator
        def wrapper(*args, **kwargs): # wrapper function to handle arguments
            for _ in range(n): # repeat n times
                func(*args, **kwargs) # call the original function
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("Hello")
```

```python
say_hello()
```

#### Conclusion
Decorators are a powerful tool in Python for extending and modifying the behavior of functions and methods. They provide a clean and readable way to add functionality such as logging, timing, access control, and more without changing the original code.
